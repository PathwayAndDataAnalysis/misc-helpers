```python
# Some of rows has multiple gene symbols. So for now to make simple deleting all the rows which has multiple gene symbols
# Multiple genes has some relationship with the same peptides. 

# Removing rows which has multiple gene names
index_to_delete = []
for index, row in df1_copy.iterrows():
    # Access any cell in row and set it to 0
    # Check if value in cell fulfils condition
    if len(row['Gene']) > 1:
        index_to_delete.append(index)

df1_copy.drop(df1_copy.index[index_to_delete], inplace=True)

df1_copy
```

```python
# df1_copy['Gene'].to_csv("GenesFinal.csv", index=False)
```

```python
# Reset indexes
df1_copy.reset_index(drop=True, inplace=True)
df1_copy
```

```python
# Removing rows which is empty gene name
for index, row in df1_copy.iterrows():
    if not row['Gene']:
        df1_copy.drop(df1_copy.index[index], inplace=True)

df1_copy
```

```python
# Reset index of dataframe
df1_copy.reset_index(drop=True, inplace=True)
df1_copy
```

##### Reducing from list of genes to gene symbols and putting into 'Gene' Column

```python
df1_copy2 = df1_copy.copy()

for index, row in df1_copy2.iterrows():
    # Access any cell in row and set it to 0
    # Check if value in cell fulfils condition
    for item in row['Gene']:
        if item.startswith("GN="):
            df1_copy2.loc[index, 'Gene'] = item[3:]
    else:
        pass

# Deleting splitted column
df1_copy2.drop('masterProteinSplit', axis=1, inplace=True)

df1_copy2
```

```python
df1_copy2['Gene'].to_csv("final gene.csv", index=False)
```

## Filtering out Sites


```python
df1_copy['modProteinSplit'].to_csv('sites.csv', index=False)
```

##### Filter sites

```python
df1_copy3 = df1_copy2.copy()

df1_copy3['Sites'] = df1_copy3['modProteinSplit'].apply(
    lambda x: [
        str(i) if '-' not in i else "" for i in x
    ]
)

df1_copy3
```

```python
# Removing empty elements from the sites list
df1_copy3['Sites'] = df1_copy3['Sites'].apply(
    lambda x:  [i for i in x if i != '']
)

df1_copy3
```

```python
df1_copy3['Sites'].to_csv("sites_filtered.csv", index=False)
```

```python
# Few rows has no sites, so deleting all the rows which does not have sites
index_to_delete = []
for index, row in df1_copy3.iterrows():
    if not row['Sites']:
        index_to_delete.append(index)

df1_copy3.drop(df1_copy3.index[index_to_delete], inplace=True)

df1_copy3
```

```python
df1_copy3.reset_index(drop=True, inplace=True)
df1_copy3
```

```python
df1_copy3['Sites'].to_csv("sites_filtered_cleaned.csv", index=False)
```

```python
# Extracting only sites from Sites column
df1_copy4 = df1_copy3.copy()

for row_num, row in df1_copy4.iterrows():
    sites = ''
    for item in row['Sites']:
        for index, element in enumerate(item):
            if element == '[':
                if sites == '':
                    sites = item[index:]
                else:
                    sites = sites + '; ' + item[index:]
                break
        df1_copy4.at[row_num,'Sites'] = sites

df1_copy4
```

```python
df1_copy4['Sites'].to_csv("sites_filtered_cleaned.csv", index=False)
```

```python
# Removing duplicate sites
df1_copy5 = df1_copy4.copy()

for row_num, row in df1_copy5.iterrows():
    splitted_sites = row['Sites'].split('; ')
    unique_list = []
    for item in splitted_sites:
        item = item.replace("[", "")
        item = item.replace("]", "")
        unique_list.append(item)
    df1_copy5.at[row_num,'Sites'] = list(set(unique_list))

df1_copy5
```

```python
df1_copy5['Sites'].to_csv("sites_filtered_cleaned.csv", index=False)
```

```python
# Using reg ex to remove parenthesis and value inside the parenthesis
import re
pattern = "\([^()]*\)"

df1_copy6 = df1_copy5.copy()

for row_num, row in df1_copy6.iterrows():
    site = ''
    for item in row['Sites']:
        if '/' in item:
            continue
        result = re.sub(pattern, '' ,item)
        if site == '':
            site = result
        else:
            site = site + "|" + result

    df1_copy6.at[row_num, 'Sites'] = site

df1_copy6
```

```python
df1_copy6['Sites'].to_csv("sites_finalized.csv", index=False)
```

```python
# After analyzing the sites few rows doesnot have sites. So lets remove those rows
# Count NaN values in "Master Protein Descriptions" column
df1_copy6['Sites'].replace('', np.nan, inplace=True)
print(df1_copy6['Sites'].isna().sum())

df1_copy6 = df1_copy6[df1_copy6['Sites'].notna()]
df1_copy6
```

```python
# Reset index of dataframe
df1_copy6.reset_index(drop=True, inplace=True)
```

```python
# Finally deleting un-necessary modProteinSplit column
df1_copy6.drop('modProteinSplit', axis=1, inplace=True)

df1_copy6
```

#### Creating SignedP Value Column
