```python
df1 = pd.read_excel('#1 EC_117-M982-A01-B04-PO89-P11854-1-pSTY-PepNorm-SumInd-20210917-pep-grps-HCD-MS2-Quantitation.xlsx')
df1
```

```python
df1_copy = df1.copy()
df1_copy['split'] = df1['Master Protein Descriptions'].str.split(" ")
# df1_copy
```

```python
# Checking nan values
df1_copy['Master Protein Descriptions'].isna().sum()
```

```python
df1_copy = df1_copy[df1_copy['Master Protein Descriptions'].notna()]
df1_copy
```

```python
genes = []

for index, row in df1_copy.iterrows():
    # print(row["split"])
    (row['split']).reverse()
    for i in row['split']:
        if i.startswith("GN"):
            genes.append(i)
            print(i)
            break
        

gene_series = pd.Series(genes)
```

```python
gene_series
```

```python
df1_copy['genes'] = gene_series
df1_copy
```
