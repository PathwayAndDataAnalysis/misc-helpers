```python
# Import all the necessary libraries
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import requests
import os
import pandas as pd
import numpy as np
import warnings

# Do not print warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_excel(
    '#1 EC_117-M982-A01-B04-PO89-P11854-1-pSTY-PepNorm-SumInd-20210917-pep-grps-HCD-MS2-Quantitation.xlsx')
df1
```

```python
# Count NaN values in "Master Protein Descriptions" column
df1['Master Protein Descriptions'].isna().sum()
```

```python
# Remove all the rows that contains NaN values in "Master Protein Descriptions" column
df1 = df1[df1['Master Protein Descriptions'].notna()]
df1
```

```python
# Count NaN values in "Modifications in Proteins" column
df1['Modifications in Proteins'].isna().sum()
```

```python
# Remove all the rows that contains NaN values in "Modifications in Proteins" column
df1 = df1[df1['Modifications in Proteins'].notna()]
df1
```

```python
# Spliting "Master Protein Description" column with space(" ") character
df1['masterProteinSplit'] = df1['Master Protein Descriptions'].str.split(" ")
```

```python
# Spliting "Modifications in Proteins" column with ("]; ") character
df1['modProteinSplit'] = df1['Modifications in Proteins'].str.split("]; ")
```

```python
# Verify splitted dataframe
df1.shape
```

##### Filtering out Genes


```python
df1_copy = df1.copy()
df1_copy['Gene'] = df1_copy['masterProteinSplit'].apply(
    lambda x: [str(i) if i.startswith('GN=') else '' for i in x])
df1_copy
```

```python
df1_copy['Gene'].to_csv("GenesTest.csv", index=False)
```

```python
df1_copy['Gene'] = df1_copy['Gene'].apply(
    lambda x: ''.join([str(i) if i != "" else '' for i in x]))
df1_copy
```

##### Getting all gene symbols and putting into 'Gene' Column

```python
df1_copy2 = df1_copy.copy()

for index, row in df1_copy2.iterrows():
    # Access any cell in row and set it to 0
    # Check if value in cell fulfils condition
    for item in row['Gene']:
        if item.startswith("GN="):
            df1_copy2.loc[index, 'Gene'] = item[3:]
            break
    else:
        pass

df1_copy2['Gene'].to_csv('GenesFiltered.csv', index=False)

# Deleting splitted column
df1_copy2.drop('masterProteinSplit', axis=1, inplace=True)

df1_copy2
```

### Filtering out Sites


```python
df1_copy['modProteinSplit'].to_csv('sites.csv', index=False)
```

```python
df1_copy3 = df1_copy2.copy()

df1_copy3['Sites'] = df1_copy3['modProteinSplit'].apply(
    lambda x: [str(i) if '-' not in i else "" for i in x])
df1_copy3
```

```python
df1_copy3['Sites'].to_csv("sites_filtered.csv", index=False)
```

```python
df1_copy3['Sites'] = df1_copy3['Sites'].apply(
    lambda x: ''.join([str(i) if i != "" else '' for i in x]))
df1_copy3
```
